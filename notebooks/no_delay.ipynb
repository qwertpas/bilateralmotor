```python
from sympy import *

x_h = symbols('x_h')
x_j, x_j_dot, x_r, x_r_dot = symbols('x_j, \dot{x}_j, x_r, \dot{x}_r')
m_h, m_r = symbols('m_h, m_r')
K_pj, K_dj, K_pr, K_dr = symbols('K_pj, K_dj, K_pr, K_dr')


# F_r = K_pr*(x_j - x_r)
# F_j_tr = K_pj*(x_h - x_j)
F_r = K_pr*(x_j - x_r) + K_dr*(x_j_dot - x_r_dot)
F_j_tr = K_pj*(x_h - x_j) + K_dj*(0 - x_j_dot)
F_j_fb = (m_h/m_r) * F_r
F_j_tot = F_j_tr + F_j_fb

x_j_ddot = F_j_tot / m_h
x_r_ddot = F_r / m_r

simplify(x_j_ddot)

A = Matrix([
    [diff(x_j_ddot, x_j_dot), diff(x_j_ddot, x_j), diff(x_j_ddot, x_r_dot), diff(x_j_ddot, x_r)],
    [1, 0, 0, 0],
    [diff(x_r_ddot, x_j_dot), diff(x_r_ddot, x_j), diff(x_r_ddot, x_r_dot), diff(x_r_ddot, x_r)],
    [0, 0, 1, 0]
])

A = simplify(A)
display(A)

eigs = A.eigenvals().keys()
for eig in eigs:
    print((eig))
    print()

```

```python
import numpy as np
subs_dict = {
    K_pj: 14000,
    K_dj: 200,
    K_pr: 1000,
    K_dr: 100,
    m_h: 3.5,
    m_r: 8.75
}
vals = []
vecs = []
for (eigenvalue, multiplicity, eigenvector) in A.eigenvects():
    val = nsimplify(eigenvalue.subs(subs_dict).n(), tolerance=1e-10, rational=True)
    vec = nsimplify(eigenvector[0].subs(subs_dict).n(), tolerance=1e-10, rational=True)
    vals.append(val)
    vecs.append(vec)
    display(val)
    display(vec)
vals = np.array(vals, dtype=complex)
vecs = np.array(vecs, dtype=complex)
```

```python
# Step response of mass-spring-damper
m, k, c = 3.5, 14000, 200
wn = np.sqrt(k/m)  # natural frequency
zeta = c/(2*np.sqrt(m*k))  # damping ratio
t = np.linspace(0, 0.5, 1000)

# Step response solution
x = 1 - np.exp(-zeta*wn*t)*(np.cos(wn*np.sqrt(1-zeta**2)*t) + zeta/np.sqrt(1-zeta**2)*np.sin(wn*np.sqrt(1-zeta**2)*t))

plt.figure(figsize=(8,4))
plt.plot(t, x)
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Position')
plt.title('Step Response')
plt.show()
```
